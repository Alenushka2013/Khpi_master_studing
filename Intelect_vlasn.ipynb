У травні 2021 року в Telegram з’явилася інформація про нібито витік персональних даних із державного застосунку «Дія». Зловмисники поширювали великий масив даних (ПІБ, телефони, дати народження, ІПН, паспортні дані), заявляючи, що він походить із державних реєстрів та «Дії».

Мінцифра заперечила, що витік стався із самої «Дії». Пізніше було встановлено, що дані походили з різних джерел, зокрема приватних сервісів, старих зливів, а не з центральної системи «Дія».

Проте незалежно від джерела витоку, випадок стосується **інформаційних правовідносин**, **захисту персональних даних**, **відповідальності за їх поширення** та **ризиків кібербезпеки**.

# **3. Висновки (структуровано для звіту)**

1. **Виникли інформаційні правовідносини**, пов’язані з обробкою, зберіганням, захистом і незаконним поширенням персональних даних громадян України.
2. **Суб’єктами** є громадяни, держава (Мінцифра), адміністратори державних реєстрів, кіберполіція та невстановлені зловмисники.
3. **Об’єктом** є персональні дані, що підпадають під правовий захист згідно із Законом «Про захист персональних даних» та міжнародними нормами.
4. **Регулювання** забезпечується національними законами (Про інформацію, Про захист персональних даних, ККУ ст. 361 та 182), а також міжнародними актами (GDPR як порівняльна практика).
5. **Ризики кібербезпеки** значні: можливість шахрайства, витоку ідентифікаційних даних, атак соціальної інженерії, порушення приватності та потенційних фінансових збитків.
6. Кейс демонструє необхідність підвищення рівня захисту державних інформаційних ресурсів, використання сучасних засобів кіберзахисту та роз'яснювальної роботи з громадянами щодо цифрової безпеки.

## **1. Опис кейсу**

У травні 2021 року у відкритому доступі з’явився великий масив персональних даних громадян України, який невідомі особи поширювали через Telegram-канали. Зловмисники заявляли, що ці дані нібито походять з державного мобільного застосунку «Дія».

Міністерство цифрової трансформації офіційно спростувало факт витоку саме з «Дії» та повідомило, що структура застосунку не передбачає централізованого зберігання копій документів. Попередній аналіз показав, що база була складена зі старих витоків різних реєстрів та приватних сервісів.

Незважаючи на це, подія стала однією з найгучніших інформаційних інцидентів в Україні й актуалізувала питання безпеки персональних даних, відповідальності за їх обробку та загроз кібербезпеці.

### **2.1. Суб’єкти правовідносин**

1. **Громадяни України** — власники персональних даних.
2. **Міністерство цифрової трансформації України** — адміністратор застосунку «Дія».
3. **Оператори державних реєстрів** (МВС, Мін'юст, ДПС).
4. **Зловмисники**, що здійснили збір і поширення даних.
5. **Платформи поширення** (Telegram).
6. **Органи розслідування** — Кіберполіція.

### **2.2. Об’єкти правовідносин**

* персональні дані: ПІБ, дата народження, адреса, ІПН, паспортні дані;
* електронні ідентифікаційні дані;
* цифрові документи;
* інформаційні ресурси державних реєстрів.

## **5. Ризики кібербезпеки**

1. **Identity theft** — оформлення кредитів, доступ до банківських сервісів.
2. **Соціальна інженерія** — фішинг, шантаж.
3. **Розширений цифровий профайлінг** — зловмисники можуть комбінувати дані з різних зливів.
4. **Довіра до державних сервісів** зменшується.
5. **Можливість атак на державні реєстри**, що зберігають критичні дані.

Кейс витоку даних, пов’язаний із застосунком «Дія», демонструє складність сучасних інформаційних правовідносин у цифровому середовищі. Основними об’єктами цих відносин є персональні дані громадян, а ключовими суб’єктами — держава, громадяни, оператори реєстрів та зловмисники.

Незаконне поширення бази персональних даних є порушенням національного законодавства України та положень міжнародних актів щодо приватності. * Держгеокадастр (оператор реєстру)

* Електронні записи про землю
* Державний інформаційний ресурс

### **Правовідносини**

* Збирання та зберігання даних
* Доступ і використання
* Захист державних інформаційних ресурсів
* Відповідальність за вторгнення в систему

Кейс показує проблеми кіберзахисту державних реєстрів та необхідність журналювання доступу, двофакторної автентифікації і постійного аудиту.

# **2. ЄС — Google Spain (C-131/12) — «Право на забуття»**

### **Суть кейсу**

Іспанський громадянин вимагав, щоб Google видалив посилання на старий борг із газети. Суд ЄС визнав, що пошуковик є «контролером даних» і повинен видаляти застарілу інформацію.

### **Суб’єкти**

* Google (контролер даних)
* Громадянин Іспанії
* Нац. агентство із захисту даних Іспанії (AEPD)

### **Об’єкти**

* Персональні дані в інтернеті
* Індекси та результати пошуку

### **Норми**

* GDPR (до його введення — Директива 95/46/EC)
* Статті про право на стирання / на приватність

### **Правовідносини**

* Обробка персональних даних пошуковими системами
* Права суб’єкта даних
* Обов’язки контролера

### **Загрози**

* Порушення приватності
* Несанкціоноване поширення застарілої інформації
* Шкода репутації

### **Значення кейсу**

Створив юридичну основу «права на забуття» в ЄС і вплинув на світову політику конфіденційності.

# **3. ЄС — YouTube & Cyando (C-682/18)**

(Відповідальність платформ)

### **Суть кейсу**

Позивачі стверджували, що YouTube та файлообмінник Cyando відповідальні за порушення авторських прав користувачами. Суд ЄС визначив: платформи не несуть прямої відповідальності, якщо активно не сприяють порушенню.

### **Суб’єкти**

* YouTube, Cyando (платформи)
* Правовласники музики/відео
* Користувачі, які завантажують контент

### **Об’єкти**

* Авторські права
* Відео- та аудіофайли

### **Норми**

* Директива 2001/29/EC
* Директива 2000/31/EC (E-Commerce)
* Директива 2019/790 (Copyright Directive)

### **Правовідносини**

* Розміщення й обробка контенту
* Відповідальність посередників
* Повідомлення та видалення (Notice & Takedown)

### **Загрози**

* Масове порушення авторських прав
* Комерційні збитки правовласників
* Можливість зловживання платформами

### **Значення кейсу**

Уточнив межі відповідальності онлайн-платформ у ЄС.

# **4. США — Viacom v. YouTube (2010–2014)**

(Застосування DMCA Safe Harbor)

### **Суть кейсу**

Viacom подала позов проти YouTube, стверджуючи, що платформа свідомо дозволяє порушення авторських прав. Суд визнав, що YouTube підпадає під «safe harbor» DMCA, якщо дотримується Notice & Takedown.

### **Суб’єкти**

* YouTube
* Viacom (MTV, Comedy Central)
* Користувачі

### **Об’єкти**

* Відеоконтент
* Авторські права
* Хостинг-послуга

### **Норми**

* DMCA §512 (Safe Harbor)

### **Правовідносини**

* Розміщення контенту користувачами
* Процедура повідомлення про порушення
* Обмеження відповідальності провайдерів

### **Загрози**

* Масові порушення авторських прав
* Збитки правовласників
* Відсутність превентивної модерації

### **Значення кейсу**

Встановив стандарт, що платформи не зобов’язані проактивно шукати порушення.

# **5. США — Capitol Records v. Vimeo (2016)**

### **Суть кейсу**

Позов проти Vimeo за розміщення музичних треків без ліцензії. Суд визнав, що Safe Harbor також поширюється і на сервіси, орієнтовані на певні типи контенту, якщо вони добросовісно реагують на Notice & Takedown.

### **Суб’єкти**

* Capitol Records
* Vimeo
* Користувачі

### **Об’єкти**

* Музичні твори
* Авторські права

### **Норми**

* DMCA §512
* Авторське право США (Title 17)

### **Правовідносини**

* Публікація та обробка контенту
* Права правовласників
* Відповідальність відеоплатформ

### **Загрози**

* Нелегальне поширення творів
* Збитки правовласників
* Недостатній контроль платформи

### **Значення кейсу**

Уточнив застосування Safe Harbor до платформ, що працюють з музикою і відео.

# **6. ЄСПЛ — Satakunnan Markkinapörssi Oy and Satamedia Oy v. Finland (2017)**

(Баланс між приватністю та свободою слова)

### **Суть кейсу**

Фінські медіа публікували податкові дані громадян. ЄСПЛ визнав обмеження цієї діяльності правомірним, оскільки приватність переважила інтереси свободи вираження.

### **Суб’єкти**

* Медіакомпанії Satamedia
* Громадяни Фінляндії
* Національні суди
* ЄСПЛ

### **Об’єкти**

* Персональні податкові дані
* Журналістські матеріали

### **Норми**

* Ст. 8 ЄКПЛ (право на приватне життя)
* Ст. 10 ЄКПЛ (свобода вираження)

### **Правовідносини**

* Обробка персональних даних у медіа
* Баланс між приватністю та свободою ЗМІ

### **Загрози**

* Порушення приватності
* Ризики соціального тиску
* Непропорційне поширення чутливих даних

### **Значення кейсу**

## 3\. Аналіз правового регулювання

[cite_start]У цьому кейсі виникають два основні типи правовідносин[cite: 19]:

## [cite_start]4. Виявлені ризики для кібербезпеки [cite: 6, 15]

На основі аналізу кейсу можна виділити такі технічні та правові ризики:

  * **Централізація доступу:** Навіть якщо дані не зберігаються в одному місці, єдина точка входу (API "Дії") стає критичною ціллю для атак.

## 6. Висновок про рівень техніки і новизну

**Рівень техніки:** за результатами пошуку в базах (WIPO, Espacenet, Google Patents) виявлені численні публікації й патентні документи, що покривають ідеї **location-based / geolocation-based encryption** і **dynamic key generation**. **Висновок щодо новизни:** сам по собі загальний концепт «шифрування із динамічними ключами, що залежать від геолокації» **не є новим** — існують опубліковані рішення і патентні заявки з такими ознаками. Тому, щоб об’єкт вважався новим, потрібно або:

* показати технічні відмінності (суттєві ознаки), які не описані в знайдених джерелах (наприклад, унікальний алгоритм побудови ключа, механізм захисту приватності локації, нова схема розподілу ключів без центрального сервера, інтеграція з певними апаратними модулями тощо); **або**
* комбінувати вже відомі елементи в незвичний технічний спосіб, що дає несподіваний технічний ефект (наприклад, геоприв’язка + псевдоаніонізація/псевдонімізація локації + протокол без викриття реальної позиції, або використання MPC для відновлення ключа без передачі координат).

# **ТЕОРЕТИЧНІ ВІДОМОСТІ**

## **1. Ліцензійний договір**

Ліцензійний договір — це угода, за якою ліцензіар надає ліцензіату право використання об’єкта авторського права або суміжних прав у визначених межах.

Правове регулювання в Україні:

* Цивільний кодекс України, ст. 1109–1112;
* Закон України «Про авторське право і суміжні права».

## **2. Види ліцензій**

**Виключна ліцензія** — ліцензіар передає право використання об’єкта лише одному ліцензіату, зобов’язуючись не надавати аналогічні права іншим.
**Невиключна (проста)** — ліцензіар може надавати аналогічні права необмеженій кількості користувачів.
**Повна** — фактична передача майнових прав.

## **3. Open source ліцензії**

Відкриті ліцензії визначають умови використання, модифікації й поширення коду.

Основні типи:

* **MIT License** — мінімальні обмеження, дозвіл модифікації та комерційного використання.
* **GNU GPL** — вимагає поширення похідних робіт за тією ж ліцензією («copyleft»).
* **Creative Commons** — призначена для контенту та документації, а не програмного коду.

# **ПОРІВНЯЛЬНИЙ АНАЛІЗ MIT / GPL / CC**

## **1. Умови поширення та модифікації**

* **MIT** — дозволено все (використання, модифікація, інтеграція у комерційні продукти), за умови збереження повідомлення про авторство.
* **GPL** — дозволено модифікацію, але похідний продукт *обов’язково* повинен також бути відкритим під GPL.
* **Creative Commons (CC BY / CC BY-SA)** — застосовується до текстів, фото, документації; модифікації дозволені, але зберігається авторство; CC BY-SA також вимагає ліцензування похідних матеріалів за тими ж умовами.

## **2. Комерційні обмеження**

* **MIT** — комерційне використання повністю дозволене.
* **GPL** — комерційне використання дозволене, але дистрибуція повинна залишатися open source.
* **CC BY / CC BY-SA** — дозволяють комерційне використання (за винятком варіантів NC).

## **3. Вимоги щодо збереження авторства**

Усі ліцензії передбачають збереження авторських атрибуцій.

## **4. Інтеграція у комерційні проєкти**

* **MIT** — максимально зручна для комерції.
* **GPL** — проблемна, оскільки «заражає» весь код GPL-обов’язками.
* **CC** — не підходить для програмного коду, але підходить для документації CallTools.ua.

````python
import pandas as pd

class DataCheck:
    def __init__(self, file_path: str):
        self.data = pd.read_csv(file_path)

    def missing_values(self):
        missing = self.data.isnull().sum()
        return missing[missing > 0]

    def detect_outliers(self, column: str):
        series = self.data[column].dropna()
        q1, q3 = series.quantile([0.25, 0.75])
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        return series[(series < lower) | (series > upper)]

    def generate_report(self):
        report = []
        mv = self.missing_values()
        if not mv.empty:
            report.append("Пропуски:\n" + str(mv))

        for col in self.data.select_dtypes(include='number').columns:
            outliers = self.detect_outliers(col)
            if len(outliers) > 0:
                report.append(f"Аномалії у стовпці {col}: {len(outliers)} значень")

        return "\n\n".join(report)
````
