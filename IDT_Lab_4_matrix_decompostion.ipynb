```python
import numpy as np
from scipy.linalg import lu

def lu_decomposition(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform LU decomposition of a matrix.

    Args:
        x (np.ndarray): The input matrix to decompose. Must be square.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]:
            The permutation matrix P, lower triangular matrix L, and upper triangular matrix U.
    """
    if x.ndim != 2 or x.shape[0] != x.shape[1]:
        raise ValueError("Input must be a square matrix.")

    # Використовуємо функцію lu з SciPy для чисельно стійкого розкладу PA=LU
    # lu повертає P, L, U
    P, L, U = lu(x)

    return P, L, U

# Приклад використання:
A = np.array([[2, 1, 1],
              [4, -6, 0],
              [-2, 7, 2]], dtype=float)

try:
    P, L, U = lu_decomposition(A)

    print("Вхідна матриця A:\n", A)
    print("\nМатриця перестановок P:\n", P)
    print("\nНижня трикутна матриця L:\n", L)
    print("\nВерхня трикутна матриця U:\n", U)

    # Перевірка: P * A == L * U (P @ A == L @ U)
    print("\nПеревірка (P @ A):\n", P @ A)
    print("\nПеревірка (L @ U):\n", L @ U)

except ValueError as e:
    print(f"Помилка: {e}")
except NameError:
    print("\nПОПЕРЕДЖЕННЯ: Для коректної роботи функції необхідно імпортувати 'from scipy.linalg import lu'.")
```

```python
import numpy as np

def qr_decomposition(x: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Perform QR decomposition of a matrix.

    Args:
        x (np.ndarray): The input matrix (A) to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray]: The orthogonal matrix Q and upper triangular matrix R.
    """
    if x.ndim != 2:
        raise ValueError("Input must be a 2D matrix.")

    # np.linalg.qr() - стандартна функція NumPy для QR-розкладу
    # Вона повертає ортогональну матрицю Q та верхню трикутну матрицю R
    Q, R = np.linalg.qr(x)

    return Q, R

# Приклад використання:
A = np.array([[12, -51, 4],
              [6, 167, -68],
              [-4, 24, -41]], dtype=float)

Q, R = qr_decomposition(A)

print("Вхідна матриця A:\n", A)
print("\nОртогональна матриця Q:\n", Q)
print("\nВерхня трикутна матриця R:\n", R)

# Перевірка: A = Q * R (Q @ R)
print("\nПеревірка (Q @ R):\n", Q @ R)

# Перевірка ортогональності Q (Q.T @ Q ≈ I)
identity_matrix = Q.T @ Q
print("\nПеревірка ортогональності (Q.T @ Q):\n", np.round(identity_matrix, decimals=10))
```

```python
import numpy as np

def eigen(x: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Compute the eigenvalues and right eigenvectors of a square matrix.

    Args:
        x (np.ndarray): The input square matrix.

    Returns:
        tuple[np.ndarray, np.ndarray]:
            The eigenvalues (1D array) and the right eigenvectors (2D array,
            where columns are the eigenvectors).
    """
    if x.ndim != 2 or x.shape[0] != x.shape[1]:
        raise ValueError("Input must be a square matrix.")

    # np.linalg.eig() - стандартна функція NumPy для обчислення власних значень та векторів
    # Повертає: (власні_значення, власні_вектори)
    eigenvalues, eigenvectors = np.linalg.eig(x)

    return eigenvalues, eigenvectors

# Приклад використання:
A = np.array([[2, -1],
              [4, -3]], dtype=float)

eigenvalues, eigenvectors = eigen(A)

print("Вхідна матриця A:\n", A)
print("-" * 30)

print("Власні значення (Eigenvalues):\n", eigenvalues)
print("-" * 30)

print("Власні вектори (Eigenvectors - стовпці матриці):\n", eigenvectors)
print("-" * 30)

# Перевірка: A * v = lambda * v
# Беремо перше власне значення (lambda1) та відповідний вектор (v1)
lambda1 = eigenvalues[0]
v1 = eigenvectors[:, 0]

print(f"Перевірка для lambda1 = {lambda1:.2f}:")
print("A @ v1:\n", A @ v1)
print("lambda1 * v1:\n", lambda1 * v1)
# Результати мають бути дуже близькі
```


**Пояснення:** Власні значення та Власні вектори

Власні значення (Eigenvalues) та власні вектори (Eigenvectors) — це фундаментальні поняття лінійної алгебри, які описують, як лінійне перетворення (матриця) впливає на певні вектори.

1. Визначення

Власний вектор $\mathbf{v}$ матриці $A$ — це ненульовий вектор, який не змінює свого напрямку після застосування лінійного перетворення (множення на матрицю $A$). Він може лише розтягуватися або стискатися.Математично це виражається рівнянням:$$A\mathbf{v} = \lambda\mathbf{v}
$$  * $\mathbf{A}$ — квадратна матриця (перетворення).
* $\mathbf{v}$ — **Власний вектор** (Eigenvector), який зберігає напрямок.
* $\mathbf{\lambda}$ — **Власне значення** (Eigenvalue), скаляр, який показує, у скільки разів вектор $\mathbf{v}$ розтягується або стискається.

#### 2\. Геометричний зміст

Уявіть, що матриця $A$ перетворює простір. Більшість векторів змінюють і свою довжину, і свій напрямок. Власний вектор — це як **"вісь"** цього перетворення: він лише масштабується, але залишається на тій самій прямій, що проходить через початок координат.

#### 3\. Практичне застосування

Власні значення та вектори мають вирішальне значення в багатьох галузях:

* **Аналіз головних компонент (PCA)**: Використовується для зменшення розмірності даних. Власні вектори вказують на напрямки найбільшої дисперсії в даних.
* **Динамічні системи**: Визначають стабільність нерухомих точок.
* **Фізика та інженерія**: Аналіз вібрацій, стійкість конструкцій, квантова механіка.
* **PageRank (Google)**: Алгоритм ранжування сторінок у Google базувався на знаходженні головного власного вектора матриці зв'язків між веб-сторінками.$$

```python
import numpy as np

def svd(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform Singular Value Decomposition (SVD) of a matrix.

    Args:
        x (np.ndarray): The input matrix (M) to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]:
            The unitary matrix U, the 1D array of singular values S (which are on the diagonal
            of the matrix Sigma), and the unitary matrix V^T (V transpose).
    """
    if x.ndim != 2:
        raise ValueError("Input must be a 2D matrix.")

    # np.linalg.svd() - стандартна функція NumPy для SVD
    # Повертає: (U, s, Vh)
    # Vh - це V^T (транспонована матриця V), відповідно до конвенції NumPy
    U, s, Vh = np.linalg.svd(x)

    return U, s, Vh

# Приклад використання:
# Неквадратна матриця 3x2
A = np.array([[1, 1],
              [0, 1],
              [1, 0]], dtype=float)

U, s, Vh = svd(A)

print("Вхідна матриця A (3x2):\n", A)
print("-" * 50)

print("Унітарна матриця U (ліві сингулярні вектори, 3x3):\n", U)
print("-" * 50)

# Сингулярні значення (S) - це 1D вектор.
# Щоб відтворити Sigma (Sigma, 3x2), його потрібно доповнити нулями.
print("Сингулярні значення s (1D вектор):\n", s)
Sigma = np.zeros(A.shape)
Sigma[:A.shape[1], :A.shape[1]] = np.diag(s)
print("\nМатриця Sigma (3x2):\n", Sigma)
print("-" * 50)

print("Унітарна матриця V^T (праві сингулярні вектори, 2x2):\n", Vh)
print("-" * 50)

# Перевірка: A = U * Sigma * V^T
print("Перевірка (U @ Sigma @ Vh):\n", U @ Sigma @ Vh)
```

```python
import numpy as np


def lu_decomposition(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform LU decomposition of a matrix with partial pivoting.

    Args:
        x (np.ndarray): The input matrix to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]:
            The permutation matrix P, lower triangular matrix L, and upper triangular matrix U.
    """
    x = x.copy().astype(float)
    n = x.shape[0]
    L = np.eye(n)
    U = x.copy()
    P = np.eye(n)

    for i in range(n):
        # Pivoting
        max_row = np.argmax(abs(U[i:, i])) + i
        if i != max_row:
            U[[i, max_row]] = U[[max_row, i]]
            P[[i, max_row]] = P[[max_row, i]]
            if i > 0:
                L[[i, max_row], :i] = L[[max_row, i], :i]

        # Elimination
        for j in range(i + 1, n):
            L[j, i] = U[j, i] / U[i, i]
            U[j] -= L[j, i] * U[i]

    return P, L, U


def qr_decomposition(x: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Perform QR decomposition of a matrix using Gram-Schmidt orthogonalization.

    Args:
        x (np.ndarray): The input matrix to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray]: The orthogonal matrix Q and upper triangular matrix R.
    """
    x = x.astype(float)
    n, m = x.shape
    Q = np.zeros((n, m))
    R = np.zeros((m, m))

    for j in range(m):
        v = x[:, j]
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], x[:, j])
            v = v - R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]
    return Q, R


def determinant(x: np.ndarray) -> float:
    """
    Calculate the determinant of a matrix via LU decomposition.

    Args:
        x (np.ndarray): The input matrix.

    Returns:
        float: The determinant of the matrix.
    """
    P, L, U = lu_decomposition(x)
    det_P = np.linalg.det(P)
    det_U = np.prod(np.diag(U))
    return det_P * det_U


def eigen(x: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Compute the eigenvalues and right eigenvectors of a matrix.

    Args:
        x (np.ndarray): The input matrix.

    Returns:
        tuple[np.ndarray, np.ndarray]: The eigenvalues and the right eigenvectors of the matrix.
    """
    vals, vecs = np.linalg.eig(x)
    return vals, vecs


def svd(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform Singular Value Decomposition (SVD) of a matrix.

    Args:
        x (np.ndarray): The input matrix to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]: The matrices U, S, and V^T.
    """
    U, S, Vt = np.linalg.svd(x)
    return U, np.diag(S), Vt


# ======= TESTING =======
if __name__ == "__main__":
    np.set_printoptions(precision=3, suppress=True)
    A = np.array([[2, 1, 1],
                  [4, -6, 0],
                  [-2, 7, 2]], dtype=float)

    print("Original matrix A:\n", A)

    # LU decomposition
    P, L, U = lu_decomposition(A)
    print("\nLU decomposition:")
    print("P =\n", P)
    print("L =\n", L)
    print("U =\n", U)
    print("Check: P@A = L@U ? ->", np.allclose(P @ A, L @ U))

    # QR decomposition
    Q, R = qr_decomposition(A)
    print("\nQR decomposition:")
    print("Q =\n", Q)
    print("R =\n", R)
    print("Check: Q@R = A ? ->", np.allclose(Q @ R, A))

    # Determinant
    print("\nDeterminant =", determinant(A))
    print("Check with numpy =", np.linalg.det(A))

    # Eigenvalues and eigenvectors
    vals, vecs = eigen(A)
    print("\nEigenvalues =", vals)
    print("Eigenvectors =\n", vecs)
    print("Check: A @ v = λv ? ->", np.allclose(A @ vecs, vecs @ np.diag(vals)))

    # SVD
    U, S, Vt = svd(A)
    print("\nSVD decomposition:")
    print("U =\n", U)
    print("S =\n", S)
    print("V^T =\n", Vt)
    print("Check: U @ S @ Vt = A ? ->", np.allclose(U @ S @ Vt, A))
```
