```python
import numpy as np
from scipy.linalg import lu

def lu_decomposition(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform LU decomposition of a matrix.

    Args:
        x (np.ndarray): The input matrix to decompose. Must be square.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]:
            The permutation matrix P, lower triangular matrix L, and upper triangular matrix U.
    """
    if x.ndim != 2 or x.shape[0] != x.shape[1]:
        raise ValueError("Input must be a square matrix.")

    #   lu  SciPy     PA=LU
    # lu  P, L, U
    P, L, U = lu(x)

    return P, L, U

#  :
A = np.array([[2, 1, 1],
              [4, -6, 0],
              [-2, 7, 2]], dtype=float)

try:
    P, L, U = lu_decomposition(A)

    print("  A:\n", A)
    print("\n  P:\n", P)
    print("\n   L:\n", L)
    print("\n   U:\n", U)

    # : P * A == L * U (P @ A == L @ U)
    print("\n (P @ A):\n", P @ A)
    print("\n (L @ U):\n", L @ U)

except ValueError as e:
    print(f": {e}")
except NameError:
    print("\n:       'from scipy.linalg import lu'.")
```

```python
import numpy as np

def svd(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform Singular Value Decomposition (SVD) of a matrix.

    Args:
        x (np.ndarray): The input matrix (M) to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]:
            The unitary matrix U, the 1D array of singular values S (which are on the diagonal
            of the matrix Sigma), and the unitary matrix V^T (V transpose).
    """
    if x.ndim != 2:
        raise ValueError("Input must be a 2D matrix.")

    # np.linalg.svd() -   NumPy  SVD
    # : (U, s, Vh)
    # Vh -  V^T (  V),    NumPy
    U, s, Vh = np.linalg.svd(x)

    return U, s, Vh

#  :
#   3x2
A = np.array([[1, 1],
              [0, 1],
              [1, 0]], dtype=float)

U, s, Vh = svd(A)

print("  A (3x2):\n", A)
print("-" * 50)

print("  U (  , 3x3):\n", U)
print("-" * 50)

#   (S) -  1D .
#   Sigma (Sigma, 3x2),    .
print("  s (1D ):\n", s)
Sigma = np.zeros(A.shape)
Sigma[:A.shape[1], :A.shape[1]] = np.diag(s)
print("\n Sigma (3x2):\n", Sigma)
print("-" * 50)

print("  V^T (  , 2x2):\n", Vh)
print("-" * 50)

# : A = U * Sigma * V^T
print(" (U @ Sigma @ Vh):\n", U @ Sigma @ Vh)
```

```python
import numpy as np


def lu_decomposition(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform LU decomposition of a matrix with partial pivoting.

    Args:
        x (np.ndarray): The input matrix to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]:
            The permutation matrix P, lower triangular matrix L, and upper triangular matrix U.
    """
    x = x.copy().astype(float)
    n = x.shape[0]
    L = np.eye(n)
    U = x.copy()
    P = np.eye(n)

    for i in range(n):
        # Pivoting
        max_row = np.argmax(abs(U[i:, i])) + i
        if i != max_row:
            U[[i, max_row]] = U[[max_row, i]]
            P[[i, max_row]] = P[[max_row, i]]
            if i > 0:
                L[[i, max_row], :i] = L[[max_row, i], :i]

        # Elimination
        for j in range(i + 1, n):
            L[j, i] = U[j, i] / U[i, i]
            U[j] -= L[j, i] * U[i]

    return P, L, U


def qr_decomposition(x: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Perform QR decomposition of a matrix using Gram-Schmidt orthogonalization.

    Args:
        x (np.ndarray): The input matrix to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray]: The orthogonal matrix Q and upper triangular matrix R.
    """
    x = x.astype(float)
    n, m = x.shape
    Q = np.zeros((n, m))
    R = np.zeros((m, m))

    for j in range(m):
        v = x[:, j]
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], x[:, j])
            v = v - R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]
    return Q, R


def determinant(x: np.ndarray) -> float:
    """
    Calculate the determinant of a matrix via LU decomposition.

    Args:
        x (np.ndarray): The input matrix.

    Returns:
        float: The determinant of the matrix.
    """
    P, L, U = lu_decomposition(x)
    det_P = np.linalg.det(P)
    det_U = np.prod(np.diag(U))
    return det_P * det_U


def eigen(x: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    """
    Compute the eigenvalues and right eigenvectors of a matrix.

    Args:
        x (np.ndarray): The input matrix.

    Returns:
        tuple[np.ndarray, np.ndarray]: The eigenvalues and the right eigenvectors of the matrix.
    """
    vals, vecs = np.linalg.eig(x)
    return vals, vecs


def svd(x: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Perform Singular Value Decomposition (SVD) of a matrix.

    Args:
        x (np.ndarray): The input matrix to decompose.

    Returns:
        tuple[np.ndarray, np.ndarray, np.ndarray]: The matrices U, S, and V^T.
    """
    U, S, Vt = np.linalg.svd(x)
    return U, np.diag(S), Vt


# ======= TESTING =======
if __name__ == "__main__":
    np.set_printoptions(precision=3, suppress=True)
    A = np.array([[2, 1, 1],
                  [4, -6, 0],
                  [-2, 7, 2]], dtype=float)

    print("Original matrix A:\n", A)

    # LU decomposition
    P, L, U = lu_decomposition(A)
    print("\nLU decomposition:")
    print("P =\n", P)
    print("L =\n", L)
    print("U =\n", U)
    print("Check: P@A = L@U ? ->", np.allclose(P @ A, L @ U))

    # QR decomposition
    Q, R = qr_decomposition(A)
    print("\nQR decomposition:")
    print("Q =\n", Q)
    print("R =\n", R)
    print("Check: Q@R = A ? ->", np.allclose(Q @ R, A))

    # Determinant
    print("\nDeterminant =", determinant(A))
    print("Check with numpy =", np.linalg.det(A))

    # Eigenvalues and eigenvectors
    vals, vecs = eigen(A)
    print("\nEigenvalues =", vals)
    print("Eigenvectors =\n", vecs)
    print("Check: A @ v = v ? ->", np.allclose(A @ vecs, vecs @ np.diag(vals)))

    # SVD
    U, S, Vt = svd(A)
    print("\nSVD decomposition:")
    print("U =\n", U)
    print("S =\n", S)
    print("V^T =\n", Vt)
    print("Check: U @ S @ Vt = A ? ->", np.allclose(U @ S @ Vt, A))
```
