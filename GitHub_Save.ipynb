```python
import nbformat
from google.colab import files

filename = "GNN_Colab_3.ipynb" # ім'я файлу, який завантажується з помилкою

nb = nbformat.read(filename, as_version=4)

# Видаляємо тільки widgets (outputs залишаємо!)
if "widgets" in nb.metadata:
    del nb.metadata["widgets"]

fixed_name = "GNN_Colab_3_fixed.ipynb" # нове ім'я
nbformat.write(nb, fixed_name)

files.download(fixed_name)
```
